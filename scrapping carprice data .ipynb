# Web scrapping 

```python
import pandas as pd
from selenium import webdriver
import time
from selenium.common.exceptions import NoSuchElementException
```

```python
driver=webdriver.Chrome(r"chromedriver.exe")
time.sleep(2)
url="https://www.olx.in/"
driver.get(url)#enter to olx.in through chrome browser
```

```python
driver.find_element_by_xpath("//a[@class='_2fitb']").click()#click on car to see results of used cars
```

```python
for i in range(20):
    try:
        driver.execute_script("window.scrollBy(0,20000)")
        time.sleep(3)
        driver.find_element_by_xpath("//button[@class='rui-39-wj rui-3evoE rui-1JPTg']").click()
        time.sleep(3)
    except NoSuchElementException as e:
        print(e)
        break
```

```python
urls=[]
for i in driver.find_elements_by_xpath("//li[@class='EIR5N']/a"):
    urls.append(i.get_attribute("href"))
```

```python
len(urls)
```

```python
brand=[]
model=[]
variant=[]
year=[]
fuel=[]
transmission=[]
km=[]
owners=[]
location=[]
price=[]
```

```python
for i in urls:
    driver.get(i)
    time.sleep(3)
    try:
        brand.append(driver.find_element_by_xpath("").text)
    except NoSuchElementException:
        brand.append('-')
    try:
        model.append(driver.find_element_by_xpath("//div[@class='_3JPEe']/div[2]/div/span[2]").text)
    except NoSuchElementException:
        model.append('-')
    try:
        variant.append(driver.find_element_by_xpath("//div[@class='_3JPEe']/div[3]/div/span[2]").text)
    except NoSuchElementException:
        variant.append('-')
    try:
        year.append(driver.find_element_by_xpath("//div[@class='_3JPEe']/div[4]/div/span[2]").text)
    except NoSuchElementException:
        year.append('-')
    try:
        fuel.append(driver.find_element_by_xpath("//div[@class='_3JPEe']/div[5]/div/span[2]").text)
    except NoSuchElementException:
        fuel.append('-')
    try:
        transmission.append(driver.find_element_by_xpath("//div[@class='_3JPEe']/div[6]/div/span[2]").text)
    except NoSuchElementException:
        transmission.append('-')
    try:
        km.append(driver.find_element_by_xpath("//div[@class='_3JPEe']/div[7]/div/span[2]").text)
    except NoSuchElementException:
        km.append('-')
    try:
        owners.append(driver.find_element_by_xpath("//div[@class='_3JPEe']/div[8]/div/span[2]").text)
    except NoSuchElementException:
        owners.append('-')
    try:
        location.append(driver.find_element_by_xpath("//span[@class='_2FRXm']").text)
    except NoSuchElementException:
        location.append('-')
    try:
        price.append(driver.find_element_by_xpath("//span[@class='_2xKfz']").text)
    except NoSuchElementException:
        price.append('-')
```

```python
df=pd.DataFrame({})
df['Brand']=brand
df['model']=model
df['variant']=variant
df['year']=year
df['fuel']=fuel
df['transmission']=transmission
df['km driven']=km
df['No of owners']=owners
df['location']=location
df['price']=price
df
```

```python
df.to_csv("olx.csv")
```

```python
brand=[]
model=[]
variant=[]
year=[]
fuel=[]
transmission=[]
km=[]
owners=[]
location=[]
price=[]


for i in urls:
    driver.get(i)
    time.sleep(3)
    try:
        brand.append(driver.find_element_by_xpath("//div[@class='vdp-head']").text)
    except NoSuchElementException:
        brand.append('-')
    try:
        model.append(driver.find_element_by_xpath("//div[@class='vdp-head']").text)
    except NoSuchElementException:
        model.append('-')
    try:
        variant.append(driver.find_element_by_xpath("//div[@class='vdp-head']").text)
    except NoSuchElementException:
        variant.append('-')
    try:
        year.append(driver.find_element_by_xpath("//ul[@class='gsc_row clearfix']/li[1]").text)
    except NoSuchElementException:
        year.append('-')
    try:
        fuel.append(driver.find_element_by_xpath("//ul[@class='gsc_row clearfix']/li[5]").text)
    except NoSuchElementException:
        fuel.append('-')
    try:
        transmission.append(driver.find_element_by_xpath("//ul[@class='gsc_row clearfix']/li[6]").text)
    except NoSuchElementException:
        transmission.append('-')
    try:
        km.append(driver.find_element_by_xpath("//ul[@class='gsc_row clearfix']/li[3]").text)
    except NoSuchElementException:
        km.append('-')
    try:
        owners.append(driver.find_element_by_xpath("//ul[@class='gsc_row clearfix']/li[4]").text)
    except NoSuchElementException:
        owners.append('-')
        
    location.append("delhi")
    try:
        price.append(driver.find_element_by_xpath("//span[@class='amount']").text)
    except NoSuchElementException:
        price.append('-')
```

```python
df=pd.DataFrame({})
df['Brand']=brand
df['model']=model
df['variant']=variant
df['year']=year
df['fuel']=fuel
df['transmission']=transmission
df['km driven']=km
df['No of owners']=owners
df['location']=location
df['price']=price
df
```

```python
df.to_csv("delhi.csv")
```

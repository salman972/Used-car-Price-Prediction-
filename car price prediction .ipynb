# Data cleaning

```python
#modify Brand,model and variant column as its so much of noise
```

```python
print(df['Brand'].nunique())
df['Brand'].value_counts()
```

```python
df['Brand']=df['Brand'].apply(lambda x:x.split(' ')[0])
```

```python
#cleaning Brand column
import re
brand=[]
for i in df['Brand']:
    x=re.sub(r'[0-9]','',i)
    brand.append(x)
len(brand)
```

```python
df['Brand']=brand#after cleaning modify values of Brand
df
```

```python
#cleaning noise of Model column
Model=[]
for i in df['model']:
    i=i.split()
    if len(i)>=3:
        Model.append(i[1])
    else:
        Model.append(''.join(i))
len(Model)
```

```python
df['Model']=Model
```

```python
df['Model']=df['Model'].replace('R','Wagon R')#replace R with Wagon R as during cleaning Wagon is deleted
df
```

```python
#cleaning unnecessary noise of variant column
Variant=[]
for i in df['variant']:
    i=i.split()
    if len(i)>=3:
        Variant.append(' '.join(i[2:]))
    else:
        Variant.append(' '.join(i))
Variant
```

```python
df['Variant']=Variant
df
```

```python
#Cleaning noise in Km column
Km=[]
for i in df['km driven']:
    i=re.sub(r'[A-Za-z]+','',i)
    i=re.sub('\s+','',i)
    Km.append(i)
Km
```

```python
df['Km Driven']=Km
df
```

```python
df['Km Driven']=df['Km Driven'].apply(lambda x:int(x.replace(',','')))#convert Km driven column to numeric
df
```

```python
df=df.drop(['model','variant','km driven'],axis=1)#As we create new columns of model,variant and km driven,we may drop that columns.
df
```

```python
df['No of owners'].value_counts()
```

```python
df['No of owners'].unique()
```

```python
#As we clearly see that 1st,1st owner and First owner represent single value.we have to make them one.
df['No of owners']=df['No of owners'].replace('1st','First Owner').replace('1st Owner','First Owner').replace('2nd','Second Owner').replace('2nd Owner','Second Owner').replace('3rd','Third Owner').replace('4th','Fourth & Above Owner')
df
```

```python
df['location'].nunique()
```

```python
df['location'].value_counts().head(20)
```

```python
#cleaning noise of Location column
location=[]
for i in df['location']:
    i=i.split()
    if len(i)>2:
        location.append(i[-2])
    else:
        location.append(''.join(i))
location
```

```python
df['Location']=location
df['Location']=df['Location'].apply(lambda x:x.replace(',',''))
df['Location'].value_counts()
```

```python
df['price'].unique()
```

```python
#for price column ₹ and * has no value ,we have to replace it
df['price']=df['price'].apply(lambda x:x.replace('₹ ',''))
df['price']=df['price'].apply(lambda x:x.replace('*',''))
df
```

```python

Price=[]
for i in df['price']:
    i=i.split()
    if len(i)==2:
        if i[-1]=='Lakh':
            Price.append(float(i[0])*100000)#where price value is lakh we multiply with 100000
        elif i[-1]=='Cr':
            Price.append(float(i[0])*10000000)#where price value is cr we multiply with 10000000
        else:
            pass
    else:
        Price.append(''.join(i))
len(Price)
```

```python
df['Price']=Price
df
```

```python
df=df.drop(['location','price'],axis=1)#as we have two different columns of location and price,we drop these columns
df
```

```python
df['Price']=df['Price'].apply(lambda x:str(x))#converting object type to float in price column
df['Price']=df['Price'].apply(lambda x:x.replace(',',''))
df['Price']=df['Price'].apply(lambda x:float(x))
df
```

```python
#divide dataset in independent and dependent features
x=df.drop('Price',axis=1)
y=df['Price']
```

```python
#choosing best random_state
maxacc=0
maxrs=0
lr=LinearRegression()
for i  in range(1,100):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=i)
    lr.fit(x_train,y_train)
    y_pred=lr.predict(x_test)
    acc=r2_score(y_test,y_pred)
    if acc>maxacc:
        maxacc=acc
        maxrs=i
print('best accuracy score is',maxacc,'on random_state',maxrs)
```
